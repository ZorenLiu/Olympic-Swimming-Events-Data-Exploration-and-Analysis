## **2. &emsp; Introduction**

**Ancient Olympics**

The Olympic Games have roots in the Ancient Greece. The games were held as religious festivals every four years at the heart of the Greek civilisation in Olympia. During the Games, an agreement was enacted to cease all wars as a religious tribute paid to Zeus, the King of the Greek Gods (Young, 2021). 

The first ever Olympic Games was held in 776 BC (Penn Museum, 2020). It consisted of a series of atheletic events, including a footrace of 200 yards (approx. 183 metres). Coroebus of Elis, a Greek cook and baker, was the first recorded winner. Leonides of Rhodes, one of the greatest runners during his time, won all three categories of footraces (stadion, diaulos, and hoplitodromos) over four consecutive Olympiads (164 BC to 152 BC), a total of twelve individual winnings in his career. This historical record remained unbeaten until Michael Phelps in 2016 (Stefani, 2020). Over the centries, as more atheletes attended from all around the region (such as the Spartans), more events were added, including jumping, javelin, discus throwing and wrestling. At the height of its popularity, more than 40,000 spectators attended the Olympic Games. The final Ancient Olympic Games in the recorded history was in 393 AD, during the reign of the Roman emperor Theodosius II, possibly due to a fire that destroyed the ancient stadium in which the games were held (Benagh, n.d.).

**Modern Olympics**

Many people and countries have attempted to recreate the Olympic games in the modern history, but credits have mainly been given to the French baron Pierre de Coubertin, who was the architect and founder of the International Olympic Committee (IOC) in 1894 (Young, 2021). Coubertin established the Games to be an international event that was held in a different country every four years. 

The first Olympic Games organised by the IOC was held in Athens in 1896, during which 241 atheletes from 14 countries attended. Since, the Olympics have been held every four years, with the exception of 1916 (World War I), and 1940 & 1944 (World War II). The Games have only grown in size and variety over the decades. The Winter Olympics was introduced in 1924, to incorporate events that are not possible to feature in the Summer Olympics. In 1948, the Parallel Olympics (Paralympics) was introduced, which aimed to promote the awareness and rehabilitation of veterans after World War II. In 2010, the Youth Olympics was conceived for atheletes from 14 to 18 years of age.

In the Tokyo Summer Olympics of 2020 (actually held in 2021 due to Covid-19 pandemic), 11,656 atheletes from 205 nations competed in 339 events across 33 sports.

**Swimming in the Olympics**

Swimming as a sport have been an integral event of the Olympics since the very first modern Olympics in 1896.

The games today feature four core styles of swimming strokes: *freestyle*, *backstroke*, *breaststroke* and *butterfly*. Each stroke style is different across their starting positions, movements and muscle usage. The *relay* is a team event in which four swimmers take turns to perform their individual leg. The *medley*, first introduced in 1964, is a combination of the four core styles in a single race, either swum by an individual swimmer or in a *medley relay* by a team of four. The 2020 Olympics is the first to introduce a mixed-gender event in swimming, for the *mixed medley relay*. For each event, there are sub-categories for different swimming distances, from 50 metres to 1500 metres.

However, there are a few actions to be taken to clean the data:

* The `Rank` column is of type float, it should be an integer
* Some names in the `Name` column has a space character at the end, need to be removed
* The time columns (`Split Time` and `Total Time`) are of type object/string, in the format of `MM:SS.ff`, where M is minutes, S is seconds and f is microseconds, they could be converted to type datetime
* Additionally, `Split Time` has 'none' in cells and will be replaced with a standard time 'zero'
* The individual csv file has no information on the Olympic year nor gender, a column for each should be added
* Finally, all csv data should be combined together for easier access

Let's clean the data.

&emsp;

```python
def str2time(str_time):
    """Function to convert time from dtype string to dtype datetime"""

    # If input string is 'none' then assign a datetime of zero, else return datetime
    if str_time == 'None': return datetime.strptime('00:00.000000', '%M:%S.%f') 
    return datetime.strptime(str_time, '%M:%S.%f')

def remove_space(str_name):
    """Function to remove space character at the end of names"""
    
    if str_name[-1] == ' ': return str_name[:-1]
    else: return str_name

# Initialise Olympics results dictionary, to which the Olympics results will be appended to
olympic_results = {}

# Iterate through gender and Olympic years to access individual csv files
for gender in ['Men', 'Women']:
    for year in range(1980, 2021, 4):
        
        # Construct csv file name
        file_name = 'olympic_data/' + str(year) + '_Olympics_' + gender + '.csv'
        
        # Read csv through Pandas
        df = pd.read_csv(file_name, index_col = 0)

        # Add columns for Olympic year and gender
        df['Year'] = str(year)
        df['Gender'] = gender
        
        # Append dataframe to each other
        if gender not in olympic_results.keys():
            olympic_results[gender] = df
        else:
            olympic_results[gender] = pd.concat([olympic_results[gender], df], ignore_index=True)

# Concatenate Olympic data for both genders into a single dataframe
olympic_results = pd.concat([olympic_results['Men'], olympic_results['Women']], ignore_index=True)

# Convert time columns from dtype Object to dtype datetime using function defined above
olympic_results['Split Time'] = olympic_results['Split Time'].apply(str2time)
olympic_results['Total Time'] = olympic_results['Total Time'].apply(str2time)

# Remove empty space at the end of names
olympic_results['Name'] = olympic_results['Name'].apply(remove_space)

# Rearrange dataframe columns for easier viewing
olympic_results = olympic_results[[
                                'Year',
                                'Gender',
                                'Event',
                                'Distance',
                                'Stage',
                                'Rank',
                                'Name',
                                'Split Time',
                                'Total Time',
                                'Nation',
                                'Notes']]

olympic_results = olympic_results.astype({'Rank': 'int'})
```

Several items that need to be addressed are:

* The population quantity is broken up into separate biomes, the total population quantity as well as total coastal population quantity need to be extracted
* The 'INCOMEGROUP' column will be converted to a numerical value: Low - 1, Low Middle - 2, Upper Middle - 3 and High - 4. Additionally, whether a country is a member of OECD will be ignored, and countries that are 'Not Classified' will be assigned a value of 1

Let's clean and filtre the data to only obtain the useful data.

&emsp;

```python
# Read csv
pop_data = pd.read_csv('nagdc-population-landscape-and-climate-estimates-version-3.csv')

# Rename column name
pop_data = pop_data.rename(columns = {'COUNTRYNAME': 'Nation'})

# Total population grouped by country
total_pop_df = pop_data.groupby('Nation').sum()
total_pop_df = total_pop_df[['POPULATION1990', 'POPULATION2000', 'POPULATION2010']]
total_pop_df.reset_index(level=0, inplace=True)

# Coastal population grouped by country
# Filtre dataframe to extract rows where the biome is within 100km of the coast
coastal_pop_df = pop_data[pop_data['Theme: Variable'].str.contains('COASTAL PROXIMITY: 100KM ZONE', case = False)]
coastal_pop_df = coastal_pop_df.groupby(['Nation']).sum()
coastal_pop_df = coastal_pop_df[['POPULATION1990', 'POPULATION2000', 'POPULATION2010']]
coastal_pop_df.reset_index(level=0, inplace=True)

# Rename coastal population column names
coastal_pop_df = coastal_pop_df.rename(columns = {'POPULATION1990': 'COASTALPOPULATION1990',
                                                  'POPULATION2000': 'COASTALPOPULATION2000',
                                                  'POPULATION2010': 'COASTALPOPULATION2010'
                                                  })

# Convert national income group to numeric value
income_df = pop_data[['Nation', 'INCOMEGROUP']].drop_duplicates()
income_df['INCOMEGROUP'] = income_df['INCOMEGROUP'].map({'Low income': 1,
                                                         'Lower middle income': 2,
                                                         'Upper middle income': 3,
                                                         'High income: nonOECD': 4,
                                                         'High income: OECD': 4,
                                                         'Not Classified': 1,
                                                         })

# Final popultaion dataframe
pop_df = pd.merge(total_pop_df, coastal_pop_df, on = 'Nation')
pop_df = pd.merge(pop_df, income_df, on = 'Nation')

# Calculate coastal population as a percentage of total population by year
pop_df['COSTAL1990PCT'] = pop_df['COASTALPOPULATION1990'] / pop_df['POPULATION1990']
pop_df['COSTAL2000PCT'] = pop_df['COASTALPOPULATION2000'] / pop_df['POPULATION2000']
pop_df['COSTAL2010PCT'] = pop_df['COASTALPOPULATION2010'] / pop_df['POPULATION2010']
```

**Q3. &emsp; Is there a correlation between a country's number of Olympic medals and the percentage of that country's population who live within 100km of the coast?**

Intuitively, the answer might be a yes. It might be 'obvious' that the greater percentage of a country's population that live near the coast, the more likely that country has more people with higher aptitude for swimming, hence that country will more likely to have stronger swimmers due to the larger talent pool. How strong is this connection? Let us explore this hypothesis.

&emsp;

```python
# Extract a dataframe of countries and their total Olympic medal count (Gold, Silver and Bronze)
# Extract a dataframe of percentage of a country's population that lives within 100km of the COSTAL1990PCT
# Merge the two dateframes
medals_by_nation = all_medals_df.groupby(by='Nation').size().sort_values(ascending = False).reset_index(name = 'Olympic Medals')
pop_df['AVG_PCT'] = pop_df[['COSTAL1990PCT', 'COSTAL2000PCT', 'COSTAL2010PCT']].mean(axis = 1)
medals_by_nation = pd.merge(medals_by_nation, pop_df[['Nation', 'AVG_PCT']], on = 'Nation', how = 'inner').reset_index(drop = True)

fig, axes = plt.subplots(nrows = 2, ncols = 1)

# Create bar chart
medals_by_nation.plot(
    kind = 'bar',
    title = 'Nations by Number of Olympic Medals',
    x = 'Nation',
    y = 'Olympic Medals',
    xlabel = 'Nations',
    ylabel = 'Number of Olympic Medals',
    figsize = [18, 10],
    ax = axes[0]
    )

# Create bar chart
medals_by_nation.plot(
    kind = 'bar',
    title = 'Nations by Percentage of Costal Population to Total Population',
    x = 'Nation',
    y = 'AVG_PCT',
    xlabel = 'Nations',
    ylabel = 'Percentage Costal Population',
    figsize = [18, 10],
    ax = axes[1]
    )

fig.tight_layout()
```

&emsp;

Oh no, this doesn't look good for our original hypothesis. While the first bar chart is ordered from the highest medal count to the lowest, the second bar chart, whose x-axis is arranged in the same order as the first, does not follow the same trend. In fact, there seems to be no common pattern between medal count and percentage of coastal population.

Let us look at the correlation and covariance between those two variables.

&emsp;

```python
var_corr = medals_by_nation['Olympic Medals'].corr(medals_by_nation['AVG_PCT'])
var_cov = medals_by_nation['Olympic Medals'].cov(medals_by_nation['AVG_PCT'])

print("Correlation: {:.3f}".format(var_corr))
print("Covariance: {:.3f}".format(var_cov))
```

&emsp;

The extremely small correlation of 0.029 reinforces the finding that there is very little connection between a country's performance in swimming at the Olympics and that country's proportion of coastal population.

A potential explanation is that, people do not necessarily have to go swimming in the ocean, but can also swim in a pool or at an acquatic centre. Swimming in a controlled and comfortable environment most likely yield better performance results than swimming in the outdoors. In fact, the degree of access to good public recreational facilities might trend better with medal count. This is suggested considering that 7 out of the top 10 countries by medal count are also in the top 10 countries ranked by GDP (Worldometer, n.d.). As richer countries typically have better public infrastructure. A future study will take a closer look at a country's Olympic performance to their economic performance.

**Q4. &emsp; What are the differences in time between the different swimming events?**

Let us examine the spread of times for each swimming event. A box plot can be used to compare the different events, which also shows statistical summary for each event.

&emsp;

```python
# Set absolute time zero
time_zero = datetime(1900, 1, 1)

# Extract swimming events types and recorded time, and calculate swim time in seconds
temp_df = olympic_results[['Event', 'Distance', 'Total Time']].copy()
temp_df.loc[:, 'Time Secs'] = temp_df.loc[:, 'Total Time'].apply(lambda n: (n - time_zero).total_seconds())
temp_df.loc[:, 'Event_Distance'] = temp_df.loc[:, 'Event'] + '_' + temp_df.loc[:, 'Distance']

# Order the events by median swim time in ascending order
order_list = list(temp_df.groupby('Event_Distance').median().sort_values(by='Time Secs').index)

# Create categorical order for swim events
temp_df['Event_Distance'] = pd.Categorical(
    temp_df['Event_Distance'], 
    categories = order_list, 
    ordered=True
)

# Create box plots
plt.figure(figsize=(20, 10))
plt.xticks(rotation=45)
ax = sns.boxplot(x = 'Event_Distance', y = 'Time Secs', data = temp_df)
ax.set_title('Olympic Swimming Events Comparison by Swim Time')
ax.set_xlabel('Swimming Events')
ax.set_ylabel('Time (seconds)')
```

```python
temp_df.groupby('Event_Distance').describe()
```

&emsp;

In the figure above, it is obvious that events with shorter distances have lower swim time spread. We can also observe the order of the four stroke types by speed from the fastest stroke to the slowest: freestyle, butterfly, backstroke, breaststroke, whether we are looking at the mean or the lowest time.

**Q6. &emsp; Is there a significant difference in an athelete's performance between an individual event and a relay team event?**

Let us first explore the motivation.

In a relay swim, a common strategy is to order the four swimmers in the following way: second-fastest, third-fastest, slowest, and the fastest (the anchor). The fastest swimmer always swims last as they would be best able to make up the lost time by previous teammates. However, assuming if everyone on the team is swimming as fast as they possibly can, which is surely the case especially at a final in the Olympic games, why does order matter? In the end, the total is simply the summation of the four legs, doesn't matter if the fatest time is the first or last in the addition.

Hence, in this question, I want to find out how does an athelete's performance, in terms swim time, differs between at an individual event and at a relay event. Additionally, I will take into account the difference in performance across the four legs of a relay. Note, by difference in performance, it means we will look at the difference in swim time across both events in seconds.

To answer this question, we need to standardise our measures, hold as many parameters constant as we can.

Firstly, there are many types strokes in individual events, whereas the most popular relay event is the freestyle. From the answer to Q4, we know that on average there is a difference between the swim times of different strokes. Hence, we will only compare individual event and relay event of the same stroke, which will be the freestyle.

Secondly, as we have also seen from Q4, the distance of an event also has a significant impact on time, hence, we will only consider races where the atheletes swim the same distance in their individual legs. Hence, we will only consider the 100m freestyle, and the 4 x 100m freestyle relay.

Thirdly, because there can be a great contrast in performance between different atheletes, we will only consider an athelete who has competed both individually and in a relay event.

Finally, as we saw in Q5, there is a general downward trend of faster swims over the Olympic years. Also, an indivudal athelete's perform may vary Olympic year to Olympic year. Hence, we will only consider compare an athelet's performance in the same Olympic year.

With those four considerations, let us start to extract the necessary data.

&emsp;

```python
from itertools import cycle

# We will extract two dataframes, one for freestyle and one for freestyle relay

# Consideration 1 and 2: extract only 4 x 100m freestyle relay event
relay_df = olympic_results[(olympic_results['Event'].str.contains('Freestyle relay')) & (olympic_results['Distance'].str.contains('100'))].copy()
# Create the leg order number for each athelete
pattern = cycle([1, 2, 3, 4])
relay_df['Relay Leg'] = [next(pattern) for pat in range(len(relay_df))]
# Consideration 3 and 4: group by Olympic year and athelete
relay_df = relay_df.groupby(['Year', 'Name']).min()
relay_df = relay_df[['Split Time', 'Total Time', 'Relay Leg']]
relay_df = relay_df.reset_index()

# Consideration 1 and 2: extract only 100m freestyle relay event
freestyle_df = olympic_results[(olympic_results['Event'].str.contains('Freestyle')) & (olympic_results['Distance'].str.contains('100')) & ~(olympic_results['Event'].str.contains('relay'))].copy()
freestyle_df = freestyle_df.rename(columns = {'Total Time': 'Indi Event Time'})
# Consideration 3 and 4: group by Olympic year and athelete
freestyle_df = freestyle_df.groupby(['Year', 'Name']).min()
freestyle_df = freestyle_df['Indi Event Time']
freestyle_df = freestyle_df.reset_index()

# Now we merge both dataframes to only compare atheletes who have both competed in an individual event and a relay event
cmb_df = pd.merge(freestyle_df, relay_df, on = ['Year', 'Name'], how = 'inner')

# Calculate the difference in performance
cmb_df['Difference'] = (cmb_df['Split Time'] - cmb_df['Indi Event Time']).dt.total_seconds()

# We split the dataframe into their individual relay legs
order_1_df = cmb_df[cmb_df['Relay Leg'] == 1]
order_2_df = cmb_df[cmb_df['Relay Leg'] == 2]
order_3_df = cmb_df[cmb_df['Relay Leg'] == 3]
order_4_df = cmb_df[cmb_df['Relay Leg'] == 4]
```

&emsp;

Now we have cleaned the data, let us visualise the results. We will use box plots to visualise the spread in the difference in performance. Note, due to the way we calculated the difference, a negative difference means the athelete had a faster time in the relay event, and a positve difference means the athelete had a faster time in the individual event.

&emsp;

```python

# Create the subplots plots
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize = (12,5), sharey = True)

# Create the boxplots for each subplot
sns.boxplot(y = 'Difference', data = order_1_df,  orient = 'v' , ax = axes[0]).set(ylim=(-4, 4))
sns.boxplot(y = 'Difference', data = order_2_df,  orient = 'v' , ax = axes[1]).set(ylim=(-4, 4))
sns.boxplot(y = 'Difference', data = order_3_df,  orient = 'v' , ax = axes[2]).set(ylim=(-4, 4))
sns.boxplot(y = 'Difference', data = order_4_df,  orient = 'v' , ax = axes[3]).set(ylim=(-4, 4))

# Add figure super title
fig.suptitle('Time Difference For Each Leg in a 4 x 100m Relay')

# Add individual xlabel
for n in range(4):
    xlab = 'Relay Leg ' + str(n+1)
    axes[n].set_xlabel(xlab)
    
    if n == 0:
        axes[n].set_ylabel('Time Difference (seconds)')
    else:
        axes[n].set_ylabel('')

fig.tight_layout()
```

The order of the performance increase to leg order from greatest to least is: leg 3, leg 2, leg 4, and leg 1.

At the performance level of the Olympic games, even 0.1 of a second is the difference between Gold and silver. As a scale reference, the mens 4 x 100m freestyle relay final at the Tokyo Olympics, Australia came third, 0.11 seconds behind second place, and 1.25 seconds behind first place.

However, by analysis these numbers will not tell us precisely why an athelete seems to swim faster if they race in a relay, and significantly so if they swim in particular leg of a relay. We may need to  conduct controlled experiements to find out. However, I postulate, the psychology of the athelete differs between an individual event and a relay. Typically, swimming is performed on an individual level, even in a relay, the athelete still needs to perform their leg independently. The psychology of competitiveness thus may become a factor. In individual events, the athelete are competing with other athelets from other countries. In relay events, on top of competing with international atheletes, they are also 'competing' against their own national atheletes, wanting to swim the fastest leg. Additionally, it is also likely that the athelete from the same nation also compete with each other in the individual events and even outside of the Olympics. Hence, the level of competitiveness can be argued to be higher than individual event, hence giving the atheletes a slight boost in their performance.

Again, this is pure postulation. More data is needed as well more experiements need to be conducted to come to a more conclusive answer.
